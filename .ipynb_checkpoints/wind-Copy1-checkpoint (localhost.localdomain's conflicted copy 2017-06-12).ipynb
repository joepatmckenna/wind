```python
# range: 1987-07-02 to 2011-12-31
# leap years: 1988, 1992, 1996, 2000, 2004, 2008
# number of days per month
# (1,31), (2,28), (3,31), (4,30), (5,31), (6,30)
# (7,31), (8,31), (9,30), (10,31), (11,30), (12,31)
# earth circumference: 40,075 km
# 0<lat<pi, 0<lon<2pi

import numpy as np
from scipy.interpolate import RectSphereBivariateSpline
from scipy.io import netcdf
import matplotlib.pyplot as plt
from matplotlib import animation
%matplotlib inline
import cartopy.crs as ccrs
from datetime import date, timedelta
from os import remove

data_root = '/home/joe/data/'
file_pre = data_root + 'podaac-ftp.jpl.nasa.gov/allData/ccmp/L3.0/flk/'

netcdf_on_local_disk = False

day = timedelta(days=1)
min_date, max_date = date(1987,7,2), date(2011,12,31)
n_days = (max_date-min_date).days
n_lat, n_lon= 628, 1440
min_lat_centered, max_lat_centered = -78.375, 78.375
min_lat, max_lat = 90-78.375, 90+78.375
min_lon, max_lon = .125, 359.875
min_lat_rad = np.pi*min_lat/180
max_lat_rad = np.pi*max_lat/180
min_lon_rad = np.pi*min_lon/180
max_lon_rad = np.pi*max_lon/180
earth_radius = 6371.008 # km
```

```python
if netcdf_on_local_disk:
    # return netcdf file for given date
    file_post = '_v11l30flk.nc'
    def date_to_cdf(d):
        filename = d.strftime('%Y/%m') + '/analysis_' + d.strftime('%Y%m%d')
        filename = file_pre + filename + file_post
        cdf_file = netcdf.netcdf_file(filename)
        return cdf_file
    def cdf_to_npz(d):
        cdf_file = date_to_cdf(d)
        uwnd = np.copy(cdf_file.variables['uwnd'].data)
        vwnd = np.copy(cdf_file.variables['vwnd'].data)
        time = np.copy(cdf_file.variables['time'].data)
        filename = file_pre + d.strftime('%Y/%m/%Y%m%d.npz')
        np.savez(filename,uwnd=uwnd,vwnd=vwnd,time=time)
        cdf_file.close()
else:
    # return lat/lon files
    def lat_lon(units='deg'):
        # lat, lon = file_pre + 'lat.npy', file_pre + 'lon.npy'
        # lat, lon = np.load(lat), np.load(lon)
        lat = 90+np.linspace(-78.375,78.375,n_lat)
        lon = np.linspace(.125,359.875,n_lon)
        if units == 'rad':
            lat = np.pi*lat/180
            lon = np.pi*lon/180
        elif units == 'km':
            lat = np.pi*earth_radius*lat/180
            lon = np.pi*earth_radius*lon/180
        return lat, lon
    # return npz files for  given date
    def date_to_npz(d, units='m_per_s'):
        filename = file_pre + d.strftime('%Y/%m/%Y%m%d.npz')
        with np.load(filename) as data:
            time,uwnd,vwnd = data['time'],data['uwnd'],data['vwnd']
            if units == 'km_per_hr':
                uwnd = 3.6*np.array(uwnd) 
                vwnd = 3.6*np.array(vwnd)
        return time, uwnd, vwnd
```

```python
# write lat and lon files if lon/lat are the same across dates
if netcdf_on_local_disk:
    current_date = min_date
    cdf_file = date_to_cdf(current_date)
    lat = np.copy(cdf_file.variables['lat'].data)
    lon = np.copy(cdf_file.variables['lon'].data)
    cdf_file.close()
    same_grid = []
    while current_date < max_date:
        current_date += day
        cdf_file = date_to_cdf(current_date)
        lat_tmp = np.copy(cdf_file.variables['lat'].data)
        lon_tmp = np.copy(cdf_file.variables['lon'].data)
        same_grid.append(all(lat==lat_tmp) and all(lon==lon_tmp))
        cdf_file.close()
    if all(same_grid):
        np.save(file_pre + 'lat.npy', lat)
        np.save(file_pre + 'lon.npy', lon)
```

```python
# convert netcdf to npz files
if netcdf_on_local_disk:
    current_date = min_date
    while current_date < max_date:
        cdf_file = date_to_cdf(current_date)
        cdf_to_npz(current_date)
        filename = file_pre + current_date.strftime('%Y/%m/%Y%m%d.npz')
        with np.load(filename) as data:
            if np.all(cdf_file.variables['time'].data == data['time']) \
            and np.all(cdf_file.variables['uwnd'].data == data['uwnd']) \
            and np.all(cdf_file.variables['vwnd'].data == data['vwnd']):
                cdf_file.close()
                remove(cdf_file.filename)
        current_date += day
    netcdf_on_local_disk = False
```

```python
# spaitial interpolate
current_date = min_date+np.random.randint(n_days)*day
time,uwnd,vwnd = date_to_npz(current_date)

lat, lon = lat_lon(units='rad')

u_spl = RectSphereBivariateSpline(lat,lon,uwnd[0])
v_spl = RectSphereBivariateSpline(lat,lon,vwnd[0])

lat, lon = np.meshgrid(lat, lon)
u = u_spl.ev(lat,lon).T
v = v_spl.ev(lat,lon).T

fig,ax = plt.subplots(2,2)
ax[0,0].imshow(uwnd[0], interpolation='nearest',origin='bottom')
ax[0,1].imshow(u, interpolation='nearest',origin='bottom')
ax[1,0].imshow(vwnd[0], interpolation='nearest',origin='bottom')
ax[1,1].imshow(v, interpolation='nearest',origin='bottom')
```

```python
ax = plt.axes(projection=ccrs.PlateCarree())
ax.coastlines()
ax.stock_img()
ax.set_global()

# interpolate
current_date = min_date+np.random.randint(n_days)*day
time,uwnd,vwnd = date_to_npz(current_date, units='km_per_hr')

lat, lon = lat_lon(units='rad')

u_spl = RectSphereBivariateSpline(lat,lon,uwnd[0])
v_spl = RectSphereBivariateSpline(lat,lon,vwnd[0])

def integrate(lat,lon):
    sin_lat, cos_lat = np.sin(lat), np.cos(lat)
    sin_lon, cos_lon = np.sin(lon), np.cos(lon)
    p = earth_radius*np.array([cos_lon*cos_lat, sin_lon*cos_lat, sin_lat])
    print np.linalg.norm(p)
    lat_tan = np.array([-cos_lon*sin_lat, -sin_lon*sin_lat, cos_lat])
    lon_tan = np.array([-sin_lon, cos_lon, 0])
    print np.linalg.norm(lat_tan), np.linalg(lon_tan)
    dp = 1*(u_spl.ev(lat,lon)*lat_tan + v_spl.ev(lat,lon)*lon_tan)
    print np.linalg.norm(dp)
    p += dp
    print np.linalg.norm(p)
    p /= np.linalg.norm(p)
    lat, lon = np.arctan2(p[1],p[0]), np.arcsin(p[2])
    lat = min(lat, min_lat_rad)
    lat = max(lat, max_lat_rad)
    lon = np.mod(lon, 2*np.pi)
    return lat, lon

p = [np.pi/2., np.pi]
traj = [p]
for i in range(6):
    p = integrate(*p)
    traj.append(p)

traj = np.array(traj)
traj *= 180/np.pi
traj[:,0] -= 90

# vector field
lat = np.linspace(min_lat_rad, max_lat_rad, 50)
lon = np.linspace(min_lon_rad, max_lon_rad, 50)
lat, lon = np.meshgrid(lat, lon)
u = u_spl.ev(lat,lon)
v = v_spl.ev(lat,lon)

lat = np.linspace(min_lat_centered, max_lat_centered, 50)
lon = np.linspace(min_lon, max_lon, 50)
lat, lon = np.meshgrid(lat, lon)
Q = plt.quiver(lon, lat, u, v, transform=ccrs.PlateCarree())

ax.plot(traj[:,1], traj[:,0], 'r-')

plt.show()

print traj
```
