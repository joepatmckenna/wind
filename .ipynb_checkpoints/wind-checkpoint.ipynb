```python
# range: 1987-07-02 to 2011-12-31
# spline: 0<lat<pi, 0<lon<2pi
# cartopy plots: -90<lat<90, -180<lon<180

import numpy as np
#from scipy.interpolate import RectSphereBivariateSpline
import matplotlib.pyplot as plt
from matplotlib import animation
%matplotlib inline
import cartopy.crs as ccrs
import cartopy.feature as cartopy_features
from datetime import date, datetime, timedelta
import wind
#print wind.rep.__doc__
#print wind.ev.__doc__
#print wind.sphere_forward_euler.__doc__
import multiprocessing as mp
import timeit

DATA_ROOT = '/home/joe/data/'
FILE_PRE = DATA_ROOT + 'podaac-ftp.jpl.nasa.gov/allData/ccmp/L3.0/flk/'

PI = np.pi
EARTH_RADIUS = 6371.008 # km
MIN_LAT, MAX_LAT = -78.375, 78.375
MIN_LON, MAX_LON = -179.875, 179.875
N_LAT, N_LON = 628, 1440

SECOND = timedelta(seconds=1)
MINUTE = timedelta(minutes=1)
HOUR = timedelta(hours=1)
DAY = timedelta(days=1)
SEC_PER_HR = HOUR.total_seconds()

MIN_DATETIME = datetime(1987,7,2)
MAX_DATETIME = datetime(2012,1,1)
N_DAYS = (MAX_DATETIME-MIN_DATETIME).days

# return lat/lon files
def lat_lon_data(units='deg'):
    lat = np.linspace(MIN_LAT,MAX_LAT,N_LAT)
    lon = np.linspace(MIN_LON,MAX_LON,N_LON)
    if units == 'rad':
        lat = PI*(lat+90)/180
        lon = PI*(lon+180)/180
    return lat, lon

def uv_data(date=MIN_DATETIME, units='m_per_s', flat=False):
    filename = FILE_PRE + date.strftime('%Y/%m/%d/') + '%02i.npz' % (date.hour/6*6,)
    with np.load(filename) as data:
        u,v = data['u'],data['v']
    if units == 'km_per_hr':
        u = 3.6*np.array(u)
        v = 3.6*np.array(v)
    return u, v

class spline(object):
    def __init__(self,x,y,z,iopt=[0,0,0],ider=[-1,0,-1,0],
                 z0=None,z1=None,s=0.):
        
        z = np.concatenate(z)
        
        nx,tx,ny,ty,c,fp,ier = wind.rep(iopt,ider,
                                        x.copy(),y.copy(),z.copy(),
                                        z0,z1,s)
        if not ier in [0, -1, -2]:
            print 'spline rep ier:', ier

        self.ier = ier
        self.fp = fp
        self.tx = tx[:nx]
        self.ty = ty[:ny]
        self.c = c[:(nx-4)*(ny-4)]
        
    def __call__(self,x,y):

        z,ier = wind.ev(self.tx,self.ty,self.c,x,y)
        
        if not ier==0:
            print 'spline ev ier:', ier
        
        self.ier = ier
        
        return z

    def ev(self,x,y):
        return self.__call__(x,y)
```

```python
lat_data,lon_data = lat_lon_data(units='rad')
# random_date=MIN_DATETIME+DAY*np.random.randint(N_DAYS)+HOUR*np.random.randint(24)
t = datetime(2005,8,14,0)
u_data,v_data = uv_data(date=t, units='km_per_hr')
u_spl = spline(lat_data,lon_data,u_data)
v_spl = spline(lat_data,lon_data,v_data)

n_lat,n_lon = 50,50
lat = np.linspace(np.deg2rad(MIN_LAT+90),np.deg2rad(MAX_LAT+90),n_lat)
lon = np.linspace(np.deg2rad(MIN_LON+180),np.deg2rad(MAX_LON+180),n_lon)
u = u_spl(lat,lon)
v = v_spl(lat,lon)

# plot
fig = plt.figure(figsize=(12,8))
ax = plt.subplot(111, projection=ccrs.PlateCarree())
# ax.add_feature(cartopy_features.OCEAN, zorder=0)
ax.add_feature(cartopy_features.LAND, zorder=0, edgecolor='k')
ax.set_global()

lat = np.linspace(MIN_LAT,MAX_LAT,n_lat)
lon = np.linspace(MIN_LON+180,MAX_LON+180,n_lon)
lon,lat = np.meshgrid(lon,lat)
mag = u*u+v*v
Q = ax.quiver(lon,lat,u,v,mag,transform=ccrs.PlateCarree())
# S = ax.streamplot(lon-180,lat,u,v,color=mag,density=4,transform=ccrs.PlateCarree())

plt.show()
```

```python
def spline_rep((x,y,z)):
    return spline(x,y,z)
                     
def spline_ev((spl,x,y)):
    return spl(x,y)

def animate_vf(min_datetime, max_datetime, dt=6*HOUR, 
               min_lat=MIN_LAT, max_lat=MAX_LAT,
               min_lon=MIN_LON, max_lon=MAX_LON,
               n_lat=50, n_lon=50, parallel=True):

    fig = plt.figure(figsize=(12,6))
    ax = fig.add_subplot(111,projection=ccrs.PlateCarree())
    ax.add_feature(cartopy_features.LAND, zorder=0, edgecolor='k')
    ax.set_global()

    datetimes = []
    t = min_datetime
    while t < max_datetime:
        datetimes.append(t)
        t += dt
    
    lat_data,lon_data = lat_lon_data(units='rad')
 
    lat_rad = np.linspace(np.deg2rad(min_lat+90),np.deg2rad(max_lat+90),n_lat)
    lon_rad = np.linspace(np.deg2rad(min_lon+180),np.deg2rad(max_lon+180),n_lon)

    lat_deg = np.linspace(min_lat,max_lat,n_lat)
    lon_deg = np.linspace(min_lon+180,max_lon+180,n_lon)

    if parallel:

        pool = mp.Pool(processes=mp.cpu_count())
        
        wind_data = np.array([uv_data(t) for t in datetimes])
        
        args = [(lat_data,lon_data,u_data) for u_data in wind_data[:,0]]
        u_spl = pool.map_async(spline_rep,args).get()
        
        args = [(lat_data,lon_data,v_data) for v_data in wind_data[:,1]]
        v_spl = pool.map_async(spline_rep,args).get()
 
        U = pool.map_async(spline_ev,[(spl,lat_rad,lon_rad) for spl in u_spl]).get()
        V = pool.map_async(spline_ev,[(spl,lat_rad,lon_rad) for spl in v_spl]).get()

        mag = U[0]*U[0]+V[0]*V[0]
        Q = ax.quiver(lon_deg,lat_deg,U[0],V[0],mag,transform=ccrs.PlateCarree())

        pool.close()
        pool.join()

    else:

        u_data,v_data = uv_data(min_datetime)

        u_spl = spline(lat_data,lon_data,u_data)
        v_spl = spline(lat_data,lon_data,v_data)
        
        u = u_spl(lat_rad,lon_rad)
        v = v_spl(lat_rad,lon_rad)

        mag = np.sqrt(u*u+v*v)
        Q = ax.quiver(lon_deg,lat_deg,u,v,mag,transform=ccrs.PlateCarree())
        
    def update_quiver(idx):

        if parallel:
            
            mag = U[idx]*U[idx]+V[idx]*V[idx]
            Q.set_UVC(U[idx],V[idx],mag)

        else:

            u_data,v_data = uv_data(datetimes[idx])
            u_spl = spline(lat_data,lon_data,u_data)
            v_spl = spline(lat_data,lon_data,v_data)
            u = u_spl(lat_rad,lon_rad)
            v = v_spl(lat_rad,lon_rad)
            mag = u*u+v*v
            Q.set_UVC(u,v,mag)
            
        return Q,
    
    plt.tight_layout()

    anim = animation.FuncAnimation(fig, update_quiver, range(1,len(datetimes)),
                                   interval=120, blit=True)

    plt.close()
    
    return anim
    
# min_datetime = datetime(2005,8,14,0)
# max_datetime = datetime(2005,9,15,0)
# anim = animate_vf(min_datetime, max_datetime, parallel=True)
# anim.save('vid/katrina_vf.mp4', dpi=100, bitrate=-1, extra_args=['-vcodec', 'libx264'])

# min_datetime = MIN_DATETIME
# max_datetime = MAX_DATETIME
# anim = animate_vf(min_datetime, max_datetime, parallel=False)
# anim.save('vid/vf.mp4', dpi=100)
# plt.close()
```

```python
def sphere_forward_euler((s0,s1,sn,ds,lat0,lon0,u_spl,v_spl)):

    return wind.sphere_forward_euler(s0, s1, sn, ds, lat0, lon0,
                                     u_spl[0].tx, u_spl[0].ty, u_spl[0].c,
                                     u_spl[1].tx, u_spl[1].ty, u_spl[1].c,
                                     v_spl[0].tx, v_spl[0].ty, v_spl[0].c,
                                     v_spl[1].tx, v_spl[1].ty, v_spl[1].c)

def integrate(t0, tn, dt=MINUTE, n_lat=10, n_lon=10,
              min_lat=MIN_LAT, max_lat=MAX_LAT,
              min_lon=MIN_LON, max_lon=MAX_LON, parallel=True):

    if parallel:
        pool = mp.Pool(processes=mp.cpu_count())
    
    lat_data,lon_data = lat_lon_data(units='rad')
    lat = np.linspace(np.deg2rad(min_lat),np.deg2rad(max_lat),n_lat)
    lon = np.linspace(np.deg2rad(min_lon),np.deg2rad(max_lon),n_lon)
    lon0,lat0 = np.meshgrid(lon,lat)

    next_data_time = datetime(t0.year,t0.month,t0.day,t0.hour/6*6)
    u_data,v_data = ['',''],['','']
    u_spl,v_spl = ['',''],['','']
    u_data[1],v_data[1] = uv_data(next_data_time, units='km_per_hr')
    u_spl[1] = spline(lat_data,lon_data,u_data[1])
    v_spl[1] = spline(lat_data,lon_data,v_data[1])

    ds = int(dt.total_seconds())
    nt = 1 + int((tn-t0).total_seconds() / ds)
    t = t0+dt

    flow = np.stack((lat0,lon0), -1)

    t_dur_sec = (tn-t0).total_seconds()
    start_time = timeit.default_timer()

    while t < tn:
        
        last_data_time = next_data_time
        next_data_time = last_data_time + 6*HOUR
        
        u_data[0],v_data[0] = u_data[1],v_data[1]
        u_data[1],v_data[1] = uv_data(next_data_time, units='km_per_hr')
        
        u_spl[0],v_spl[0] = u_spl[1],v_spl[1]
        u_spl[1] = spline(lat_data,lon_data,u_data[1])
        v_spl[1] = spline(lat_data,lon_data,v_data[1])

        s0 = int((last_data_time-t0).total_seconds())
        s1 = int((t-t0).total_seconds())
        sn = int((min(next_data_time,tn)-t0).total_seconds())
        ns = 1 + int((sn-s1)/ds)

        if parallel:
            
            args = [(s0,s1,sn,ds,
                     flow[lat_idx,lon_idx,0],
                     flow[lat_idx,lon_idx,1],
                     u_spl,v_spl) 
                    for lon_idx in range(n_lon) for lat_idx in range(n_lat)]
            data = pool.map_async(sphere_forward_euler,args)
            flow = np.reshape(data.get(),(n_lat,n_lon,2))

        else:

            for lat_idx in range(n_lat):
                for lon_idx in range(n_lon):
                    flow[lat_idx,lon_idx] = sphere_forward_euler(
                        (s0,s1,sn,ds,
                        flow[lat_idx,lon_idx,0],
                        flow[lat_idx,lon_idx,1],
                        u_spl,v_spl))                

        t += ns*dt

        print 'integrate: %0.2f%% in %0.2f min' % (100*(t-t0).total_seconds()/t_dur_sec,(timeit.default_timer()-start_time)/60.), datetime.now()
        
    if parallel:
        pool.close()
        pool.join()
        
    return flow
```

```python
def animate_lcs(min_datetime,max_datetime,dt=6*HOUR,T=3*DAY,
                n_lat=8,n_lon=8,min_lat=MIN_LAT,max_lat=MAX_LAT,
                min_lon=MIN_LON,max_lon=MAX_LON, write_dir=None, read_dir=None):
    
    if write_dir:
        with open(write_dir + 'readme.txt', 'w+') as f:
            f.write('min_datetime: %s\n' % min_datetime)
            f.write('max_datetime: %s\n' % max_datetime)
            f.write('dt: %s\n' % dt)
            f.write('T: %s\n' % T)
            f.write('n_lat: %i\n' % n_lat)
            f.write('n_lon: %i\n' % n_lon)
            f.write('min_lat: %f\n' % min_lat)
            f.write('max_lat: %f\n' % max_lat)
            f.write('min_lon: %f\n' % min_lon)
            f.write('max_lon: %f\n' % max_lon)
            
    start_time = timeit.default_timer()
    print 'animate_lcs:', datetime.now()
    
    fig = plt.figure(figsize=(12,6))
    ax = fig.add_subplot(111,projection=ccrs.PlateCarree())
    ax.add_feature(cartopy_features.OCEAN, zorder=0)
    ax.add_feature(cartopy_features.LAND, zorder=0, edgecolor='k')
    ax.set_global()

    datetimes = []
    t = min_datetime
    while t < max_datetime:
        datetimes.append(t)
        t += dt
            
    t_dur = T.total_seconds() / SEC_PER_HR

    lat = np.linspace(np.deg2rad(min_lat),np.deg2rad(max_lat),n_lat)
    lon = np.linspace(np.deg2rad(min_lon+180),np.deg2rad(max_lon+180),n_lon)
    lon2d,lat2d = np.meshgrid(lon,lat)

    lat_deg = np.linspace(min_lat,max_lat,n_lat)
    lon_deg = np.linspace(min_lon+180,max_lon+180,n_lon)

    flow = np.load(read_dir + datetimes[0].strftime('flow/%Y-%m-%d-%H.npy'))
    ftle = wind.ftle(lat2d,lon2d,flow,t_dur)
    ftle = np.roll(ftle,n_lon/2)
    ftle = np.nan_to_num(ftle)

    CMAP = plt.get_cmap('binary')
    im = ax.imshow(ftle,
                   cmap=CMAP,
                   animated=True,
                   origin='bottom',
                   extent=(min_lon,max_lon,lat_deg[1],lat_deg[-2]))

    def init():
        ftle = [[0. for _ in range(n_lon)] for _ in range(n_lon)]

        im.set_array(ftle)
 
        return im,
        
    def update_im(idx):

        t = datetimes[idx]

        if read_dir:

            flow = np.load(read_dir + datetimes[idx].strftime('flow/%Y-%m-%d-%H.npy'))
            if datetimes[idx].hour in [3,9,15,21]:
#                 I dont know why this works
                flow[...,0] = flow[...,0].T
                flow[...,1] = flow[...,1].T
            
#             ftle = calc_ftle(lat2d,lon2d,flow,n_lat=n_lat,n_lon=n_lon,T=T)
            ftle = wind.ftle(lat2d,lon2d,flow,t_dur)
            ftle = np.roll(ftle,n_lon/2)
            ftle = np.nan_to_num(ftle)
            ftle[ftle<=.1*np.max(ftle)] = None
            
        else:
            
            flow = integrate(t,t+T,n_lat=n_lat,n_lon=n_lon,
                             min_lat=min_lat,max_lat=max_lat,
                             min_lon=min_lon,max_lon=max_lon)
            
            if write_dir:
                np.save(write_dir + datetimes[idx].strftime('flow/%Y-%m-%d-%H.npy'), flow)
            
            ftle = wind.ftle(lat2d,lon2d,flow,t_dur)
            ftle = np.roll(ftle,n_lon/2)

        print 'animate_lcs: %0.2f%% in %0.2f min' % (100.*(idx+1)/len(datetimes),(timeit.default_timer()-start_time)/60.),t,datetime.now()

        im.set_array(ftle)
      
        return im,
    
    anim = animation.FuncAnimation(fig, update_im, init_func=init,
                                   frames=len(datetimes), interval=60, blit=True)

    plt.tight_layout()
    plt.close()
    
    return anim

    
min_datetime = datetime(2005,8,14,0)
max_datetime = datetime(2005,9,15,0)
anim = animate_lcs(min_datetime, max_datetime, dt=3*HOUR, n_lat=768, n_lon=768,
                   T=14*DAY, read_dir = 'vid/data/katrina_lcs/')
anim.save('vid/katrina_lcs.mp4', dpi=100, bitrate=-1, extra_args=['-vcodec', 'libx264'])
```
